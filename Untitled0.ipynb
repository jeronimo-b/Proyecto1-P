```python
import os
import re
import json
import easyocr

# ===============================
# 2. Diccionario grande de exámenes
# ===============================
RANGOS = {
    # --- Biometría Hemática ---
    "hemoglobina": (13, 16, "g/dl"),
    "hematocrito": (40, 50, "%"),
    "eritrocitos": (4.5, 6.0, "mill/µL"),
    "vcm": (80, 96, "fL"),
    "hcm": (27, 32, "pg"),
    "chcm": (32, 36, "g/dl"),
    "leucocitos": (4.0, 11.0, "K/uL"),
    "neutrofilos": (40, 70, "%"),
    "linfocitos": (20, 40, "%"),
    "monocitos": (2, 8, "%"),
    "eosinofilos": (1, 4, "%"),
    "basofilos": (0, 1, "%"),
    "plaquetas": (150, 400, "K/uL"),

    # --- Química sanguínea ---
    "glucosa": (70, 110, "mg/dl"),
    "urea": (10, 50, "mg/dl"),
    "creatinina": (0.6, 1.3, "mg/dl"),
    "acido urico": (3.5, 7.2, "mg/dl"),

    # --- Perfil lipídico ---
    "colesterol total": (0, 200, "mg/dl"),
    "colesterol hdl": (40, 60, "mg/dl"),
    "colesterol ldl": (0, 130, "mg/dl"),
    "trigliceridos": (0, 150, "mg/dl"),

    # --- Pruebas hepáticas ---
    "tgo": (0, 40, "U/L"),
    "tgp": (0, 40, "U/L"),
    "fosfatasa alcalina": (44, 147, "U/L"),
    "bilirrubina total": (0.1, 1.2, "mg/dl"),
    "bilirrubina directa": (0.0, 0.3, "mg/dl"),
    "bilirrubina indirecta": (0.2, 0.9, "mg/dl"),
    "albumina": (3.5, 5.0, "g/dl"),

    # --- Electrolitos ---
    "sodio": (135, 145, "mmol/L"),
    "potasio": (3.5, 5.0, "mmol/L"),
    "cloro": (98, 107, "mmol/L"),
    "calcio": (8.5, 10.5, "mg/dl"),
    "magnesio": (1.7, 2.2, "mg/dl"),
    "fosforo": (2.5, 4.5, "mg/dl"),

    # --- Otros comunes ---
    "proteinas totales": (6.0, 8.3, "g/dl"),
    "tsh": (0.4, 4.0, "µUI/ml"),
    "t4 libre": (0.7, 1.8, "ng/dl"),
    "t3 libre": (2.3, 4.2, "pg/ml"),
    "vitamina d": (20, 50, "ng/ml"),
    "ferritina": (30, 400, "ng/ml"),
    "hierro": (50, 170, "µg/dl")
}

# ===============================
# 3. Funciones
# ===============================

def preparar_modelo():
    """Prepara el modelo OCR."""
    return easyocr.Reader(['es'], gpu=False)

def extraer_texto(ruta, lector):
    """Extrae texto de la imagen con OCR."""
    return " ".join(lector.readtext(ruta, detail=0, paragraph=False))

def generar_patrones():
    """Genera regex para todos los exámenes del diccionario."""
    patrones = {}
    for examen in RANGOS.keys():
        patrones[examen] = re.compile(rf"{examen}[^0-9]*([0-9]+(\.\d+)?)", re.IGNORECASE)
    return patrones

def buscar_valores(texto, patrones):
    """Busca valores numéricos en el texto usando regex."""
    valores = {}
    for nombre, patron in patrones.items():
        match = patron.search(texto)
        if match:
            try:
                valores[nombre] = float(match.group(1))
            except:
                valores[nombre] = None
    return valores

def interpretar(valores):
    """Interpreta los valores con base en los rangos normales."""
    resultados = {}
    for nombre, valor in valores.items():
        if valor is not None and nombre in RANGOS:
            bajo, alto, unidad = RANGOS[nombre]
            if valor < bajo:
                resultados[nombre] = f"{valor}{unidad}: BAJO (normal {bajo}-{alto} {unidad})"
            elif valor > alto:
                resultados[nombre] = f"{valor}{unidad}: ALTO (normal {bajo}-{alto} {unidad})"
            else:
                resultados[nombre] = f"{valor}{unidad}: NORMAL"
    return resultados

def analizar_examen(ruta_imagen):
    """Pipeline completo: OCR -> valores -> interpretación."""
    lector = preparar_modelo()
    texto = extraer_texto(ruta_imagen, lector)
    patrones = generar_patrones()
    valores = buscar_valores(texto, patrones)
    analisis = interpretar(valores)

    return {
        "texto_detectado": texto,
        "valores": valores,
        "analisis": analisis
    }
```

```python
resultado = analizar_examen(ruta_imagen)
print(json.dumps(resultado, indent=2, ensure_ascii=False))
```
